```python
import numpy as np 
import pandas as pd
from scipy import stats
import researchpy as rp
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.formula.api import logit
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ool_pds = pd.read_csv('ool_pds.csv')

# PPETHM: race
#1	White, Non-Hispanic -> 0
#2	Black, Non-Hispanic -> 1
#3	Other, Non-Hispanic -> 2
#4	Hispanic -> 3
#5	2+ Races, Non-Hispanic -> 4

# W1_M1 - religion
# 1 Baptist any denomination -> 0
# 2 Atheist -> 1
# 3 Catholic -> 2
# 4 Mormon -> 3
# 5 Jewish -> 4
# 6 Muslim -> 5
# 7 Hindu -> 
# 8 Buddhist -> 7
# 9 Pentecostal -> 8
# 10 Eastern Orthodox -> 9
# 11 Other Christian -> 10
# 12 Other non-Christian -> 11
# 13 None -> 12
# 14 Refused -> 13
```

```python
df = ool_pds[["PPGENDER", "PPAGE", "PPEDUCAT", "PPETHM", "W1_M1"]]
df["EDUCBIN"] = 0
df.loc[df['PPEDUCAT'] == 4, 'EDUCBIN'] = 1
df["PPGENDER"] = df["PPGENDER"] - 1
df["PPETHM"] = df["PPETHM"] - 1
df.head()
```

```python
results = logit('EDUCBIN ~  PPGENDER', data=df).fit()
print(results.summary())
print(np.exp(results.params))
```

```python
results = logit('EDUCBIN ~  PPGENDER + C(W1_M1)', data=df).fit()
print(results.summary())
print(np.exp(results.params))
```

```python
results = logit('EDUCBIN ~  PPGENDER + C(PPETHM) + C(W1_M1)', data=df).fit()
print(results.summary())
print(np.exp(results.params))
```
